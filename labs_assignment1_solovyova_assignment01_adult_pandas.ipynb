```python
data = pd.read_csv ('adult.data.csv')
data.head()
```

# 1. Сколько мужчин и женщин (признак sex) представлено в этом наборе данных?

```python
print('Male: ', data[data['sex'] == 'Male']['sex'].count())
print('Female: ', data[data['sex'] == 'Female']['sex'].count())
```

# 2. Каков средний возраст (признак age) женщин?

```python
print('Average age of female: ', round(data[data['sex'] == 'Female']['age'].mean()))
```

# 3. Какова доля граждан Германии (признак native-country)?

```python
count = data[data['native-country'] == 'Germany']['native-country'].count()
allPeople = data['native-country'].count()

print('Count of German citizens: ', count)
print('All of people: ', allPeople)

print('Share of German citizens: ', round((100 * count) / allPeople, 2))
```

# 4-5. Каковы средние значения и среднеквадратичные отклонения возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год? 

```python
more = data[data['salary'] == '>50K']['age']
less = data[data['salary'] == '<=50K']['age']
print('Average age of people getting > 50k: ', round(more.mean(), 2))
print('mean deviation > 50: ', round(np.std(more), 2))

print('Average age of people getting <= 50k: ', round(less.mean()))
print('mean deviation <= 50: ', round(np.std(less), 2))
```

# 8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.


```python
listMaried = {'Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'}
male =data[data['sex'] == 'Male']
married = male[male['marital-status'].apply(lambda status: status in listMaried)]
unmarried = male[male['marital-status'].apply(lambda status: status not in listMaried)]
moreMarried = married[married['salary'] == '>50K']
moreUnmarried = unmarried[unmarried['salary'] == '>50K']
result = 'женатых больше' if moreMarried.shape[0] > moreUnmarried.shape[0] else 'not maried more'
result
```

```python
marriedCount = married['sex'].count()
unmarriedCount = unmarried['sex'].count()
mar = moreMarried['sex'].count()
print('доля женатых', mar/marriedCount * 100)
unmar = moreUnmarried['sex'].count()
print('доля холостых', unmar/marriedCount * 100)
```

# 9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?

```python
maxHours = data['hours-per-week'].max()
count = data[data['hours-per-week'] == maxHours]['hours-per-week'].count()
countLarge = data[(data['hours-per-week'] == maxHours) &
                  (data['salary'] == '>50K')
                 ]['hours-per-week'].count()
print('max hours:', maxHours)
print('count people: ', count)
print('countLarge: ', countLarge)
print('% countLarge: ', round(100*countLarge/count, 2))
```
